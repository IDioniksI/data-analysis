## Завдання 4
4. Обрати випадковим чином 500000 домогосподарств (без повторів 
елементів вибірки), для них обчислити середні величини усіх 3-х 
груп споживання електричної енергії, а також 

```python
def select_households_pandas(df):
    random_household = df.sample(n=500000, replace=False).sort_index()
    return random_household

def select_households_numpy(df):
    random_household = np.random.choice(df.index, size=500000, replace=False)
    random_household = df.loc[random_household].sort_index()
    return random_household
```

```python
random_household = select_households_numpy(df)
random_household
```

```python
random_household_pandas = select_households_pandas(df)
random_household_pandas
```

Порівняю швидкість виконання pandas/numpy підходів

```python
time_taken = timeit.timeit(stmt='select_households_numpy(df)', globals=globals(), number=100)
print(f"Execution time: {time_taken} seconds")
time_taken = timeit.timeit(stmt='select_households_pandas(df)', globals=globals(), number=100)
print(f"Execution time: {time_taken} seconds")
```

## Завдання 5
5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких 
основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення 
(група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

Обчислюю середні величини усіх 3-х груп споживання електричної енергії, роблю копію датасету

```python
random_household['Avg_metering'] = random_household[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean(axis=1).round(1)
random_household['Time'] = pd.to_datetime(random_household['Time'], format='%H:%M:%S')
after_18_df = random_household[random_household['Time'].dt.time > pd.to_datetime('18:00:00', format='%H:%M:%S').time()].copy()
after_18_df['Time'] = after_18_df['Time'].dt.strftime('%H:%M:%S')

after_18_df
```

```python
after_18_df = after_18_df[after_18_df['Avg_metering'] > 6]
after_18_df
```

серед відібраних визначити ті, у яких 
основне споживання електроенергії у вказаний проміжок часу 
припадає на пральну машину, сушарку, холодильник та освітлення 
(група 2 є найбільшою)

```python
after_18_df_group_2 = after_18_df[(after_18_df['Sub_metering_2'] > after_18_df['Sub_metering_1']) & (after_18_df['Sub_metering_2'] > after_18_df['Sub_metering_3'])]
after_18_df_group_2
```

а потім обрати кожен третій результат із 
першої половини та кожен четвертий результат із другої половини

```python
first_half = after_18_df_group_2.iloc[:len(after_18_df_group_2) // 2]
second_half = after_18_df_group_2.iloc[len(after_18_df_group_2) // 2:]

filtered_first_half = first_half.iloc[::3]
filtered_second_half = second_half.iloc[::4]

result_df = pd.concat([filtered_first_half, filtered_second_half])
result_df
```
